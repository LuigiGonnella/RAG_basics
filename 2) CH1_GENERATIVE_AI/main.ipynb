# GENERATOR

```python
client = Groq(api_key=os.environ['GROQ_API_KEY'])
model = "llama-3.3-70b-versatile"

def call_llm_with_full_text(input_text):
    # Join lines to obtain a single string
    text_input = '\n'.join(input_text) if isinstance(input_text, list) else input_text
    prompt = f"Please elaborate on the following content:\n{text_input}"

    try:
        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": "You are an expert in Natural Language Processing exercise professional. In your responses, do NOT cite the context provided after the sentence **augmented input context**"},
                {"role": "assistant", "content": "You can explain, read the input and answer in detail to the user query, paying attention to the provided context."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.1
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return str(e)
```

# FORMATTED RESPONSE

```python
import textwrap

def print_formatted_response(response):
    #Define the width for wrapping text
    wrapper = textwrap.TextWrapper(width=80) #Columns wide
    wrapper_text = wrapper.fill(text=response)

    #Print formatted with HEADER and FOOTER
    print("RESPONSE:")
    print("="*30)
    print(wrapper_text)
    print("="*30)
```

# RETRIEVAL METRICS

**COSINE_SIMILARITY**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_similarity(text1, text2):
    vectorizer = TfidfVectorizer(
        stop_words='english', #Ignore common english words to focus on meaningful content
        use_idf=True,
        norm = 'l2', #L2 norm to output vectors
        ngram_range=(1, 2), #unigrams and bigrams (of words, as we see in analyzer field) to contruct the VOCABULARY
        #es. ['Love', 'ML', 'Love ML']
        sublinear_tf=True, #Apply sublinear TF scaling (logarithmic TF scaling)
        analyzer='word', #word level features, we could use also 'char' or 'char_wb'
    )
    tfidf = vectorizer.fit_transform([text1, text2])
    similarity = cosine_similarity(tfidf[0:1], tfidf[1:2]) #between first argument (query) and every other element in second argument (docs)
    #written as slices so that it is always 2D,a s cosine_similarity function wants
    return similarity[0][0]
```

**ENHANCED SIMILARITY**

```python
import spacy
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
from collections import Counter
import numpy as np

# Load spaCy model
nlp = spacy.load("en_core_web_sm") #preprocess text to reduce noise and expands terms using synonym list in WORDNET

def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        #synsets cotains
        #Example for "happy":

        # wordnet.synsets("happy") returns synsets like happy.a.01, felicitous.s.01, etc.
        # Each synset's .lemmas() gives: [Lemma('happy.a.01.happy'), Lemma('happy.a.01.felicitous'), ...]
        # .name() extracts the string: "happy", "felicitous", "glad", etc.

        #where n=noun, v=verb, a =asjective etc..
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return synonyms

def preprocess_text(text):
    doc = nlp(text.lower()) #tokenize and makes available the lemma (root form)
    lemmatized_words = []
    for token in doc:
        if token.is_stop or token.is_punct: #skips common words
            continue
        lemmatized_words.append(token.lemma_)
    return lemmatized_words

def expand_with_synonyms(words):
    expanded_words = words.copy()
    for word in words:
        expanded_words.extend(get_synonyms(word)) #like append but takes each element of the appended list individually instead of appending the entire list
    return expanded_words


def calculate_enhanced_similarity(text1, text2): #with just TFs, NOT TF-IDFs
    # Preprocess and tokenize texts
    words1 = preprocess_text(text1)
    words2 = preprocess_text(text2)

    # Expand with synonyms
    words1_expanded = expand_with_synonyms(words1)
    words2_expanded = expand_with_synonyms(words2)

    # Count word frequencies
    freq1 = Counter(words1_expanded)
    freq2 = Counter(words2_expanded)

    # Create a set of all unique words across the two texts
    unique_words = set(freq1.keys()).union(set(freq2.keys()))

    # Create frequency vectors for every single word
    vector1 = [freq1[word] for word in unique_words]
    vector2 = [freq2[word] for word in unique_words]

    # Convert lists to numpy arrays
    vector1 = np.array(vector1)
    vector2 = np.array(vector2)

    # Calculate cosine similarity
    cosine_similarity = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))

    return cosine_similarity
```

# NAIVE RAG

```python
def find_best_match_keyword_search(query, db_records):
    best_score = 0
    best_record = None

    #Split query into single words
    query_words = set(query.lower().split())

    #Iterate through each record in db_records
    for record in db_records:
        #Split record
        record_words = set(record.lower().split())

        #Calculate number of common words
        n_common = len(query_words.intersection(record_words))
        
        #Update best record according to n_common
        if n_common > best_score:
            best_score = n_common
            best_record = record
    
    return best_score, best_record
```

```python
best_keyword_score, best_matching_record = find_best_match_keyword_search(query, db_records)

print(f"Best Keyword Score: {best_keyword_score}")
print_formatted_response(best_matching_record)
```

# METRICS

```python
#Cosine similarity
cosine_simil = calculate_cosine_similarity(query, best_matching_record)
print(f"Best Cosine Similarity Score: {cosine_simil:.3f}")
```

```python
#Enhanced Cosine similarity
enhanced_cosine_simil = calculate_enhanced_similarity(query, best_matching_record)
print(query,": ", best_matching_record)
print(f"Best Enhanced Cosine Similarity Score: {enhanced_cosine_simil:.3f}")
```

### VECTOR SEARCH

**SEARCH FUNCTION**

```python
def find_best_match_vector_search(text_input, records):
    best_score = 0
    best_record = None

    for record in records:
        current_score = calculate_cosine_similarity(text_input, record)

        if current_score > best_score:
            best_score = current_score
            best_record = record
    
    return best_score, best_record
```

### INDEX-BASED SEARCH

**SEARCH FUNCTION**

```python
def setup_vectorizer(records):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(records)
    return vectorizer, tfidf_matrix

def find_best_match_index_search(query, vectorizer, docs, doc_embeddings):
    query_embedding = vectorizer.transform([query])
    cosine_similarities = cosine_similarity(query_embedding, doc_embeddings)

    id_max = cosine_similarities.argmax()
    best_record = docs[id_max]
    best_score = cosine_similarities[0, id_max]
    
    return best_score, best_record
```

```python
vectorizer, tfidf_matrix = setup_vectorizer(db_records)

best_similarity_score, best_matching_record =  find_best_match_index_search(query, vectorizer, db_records, tfidf_matrix)
augmented_input = f"Here is the query: {query}\nHere is the **augmented input context** you must refer to: {best_matching_record}"
print_formatted_response(augmented_input)
```

**METRICS**

```python
# Cosine Similarity
print(f"Best Cosine Similarity Score: {best_similarity_score:.3f}")
print_formatted_response(best_matching_record)
```
